# The optimization algorithms.

In this example,
we will discover the different high-levels functions
related to optimization algorithms.


```python
from __future__ import annotations

from gemseo import get_algorithm_options_schema
from gemseo import get_available_opt_algorithms
```

## Get available optimization algorithms


The [get_available_opt_algorithms()][gemseo.get_available_opt_algorithms] function returns the list
of optimization algorithms available in GEMSEO or in external modules



```python
get_available_opt_algorithms()
```

## Get options schema

For a given optimization algorithm, e.g. `"NLOPT_SLSQP"`,
we can get the options; e.g.



```python
get_algorithm_options_schema("NLOPT_SLSQP")
```

Or import its settings model and pass it directly with the keyword
"algo_settings_model".



```python
from gemseo.settings.opt import NLOPT_SLSQP_Settings  # noqa: E402, I001

settings_model = NLOPT_SLSQP_Settings()
```

See [this page][algorithm-settings] for more information on how to use settings models.
