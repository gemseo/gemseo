# Basic history.

In this example, we illustrate the use of the [BasicHistory][gemseo.post.basic_history.BasicHistory] post-processing
on the Sobieski's SSBJ problem.

The [BasicHistory][gemseo.post.basic_history.BasicHistory] can plot any of the constraint or objective functions
w.r.t. the optimization iterations or sampling snapshots.


```python
from __future__ import annotations

from gemseo import execute_post
from gemseo.settings.post import BasicHistory_Settings
```

We can post-process the objective:



```python
execute_post(
    "sobieski_mdf_scenario.h5",
    settings_model=BasicHistory_Settings(
        variable_names=["y_4"],
        save=False,
        show=True,
    ),
)
```

We can also post-process the constraints:



```python
execute_post(
    "sobieski_mdf_scenario.h5",
    settings_model=BasicHistory_Settings(
        variable_names=["g_1", "g_2", "g_3"],
        save=False,
        show=True,
    ),
)
```
