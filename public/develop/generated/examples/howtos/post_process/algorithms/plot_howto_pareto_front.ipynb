# Pareto front

## Problem

After an optimisation or a DOE, you want to identify the trade-off between
two or more objectives and visualise the Pareto-optimal points.

## Solution

Use [ParetoFront][gemseo.post.pareto_front.ParetoFront],
which produces a matrix of scatter plots for the selected objectives.
Locally non-dominated points for each objective are shown in red;
globally Pareto-optimal points (across all objectives) are shown in green.

## Step-by-step guide


```python
from __future__ import annotations

from gemseo import execute_post
from gemseo.settings.post import ParetoFront_Settings
```

### 1. Prerequisites

This how-to requires an HDF5 file generated by a previously executed scenario.
See [How to post-process an HDF5 file][post-process-an-hdf5-file]
for more information on how to generate and load such a file.

!!! note
    The `execute_post` function also handles previously executed instances of `MDOScenario`
    or `OptimzationProblem`.



### 2. Run the Pareto front post-processing

Use `objectives` to select the functions to include in the Pareto analysis:
Use `objectives_labels` to assign readable labels to the objective components:



```python
execute_post(
    "sobieski_mdf_scenario.h5",
    settings_model=ParetoFront_Settings(
        objectives=["g_3", "-y_4"],
        save=False,
        show=True,
    ),
)
```

## Summary

[ParetoFront][gemseo.post.pareto_front.ParetoFront] identifies and visualises
trade-offs between selected objectives.
Red points are locally non-dominated for a given objective;
green points are globally Pareto-optimal across all selected objectives.
Pass any HDF5 file from a previously executed scenario to
[execute_post()][gemseo.execute_post].
