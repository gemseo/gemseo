# History plot of problem functions

## Problem

After an optimisation or a DOE, you want to plot the evolution of selected
problem functions across iterations or sampling snapshots.

## Solution

Use [BasicHistory][gemseo.post.basic_history.BasicHistory],
which plots any problem function against the iteration index.

## Step-by-step guide


```python
from __future__ import annotations

from gemseo import execute_post
from gemseo.settings.post import BasicHistory_Settings
```

### 1. Prerequisites

This how-to requires an HDF5 file generated by a previously executed scenario.
See [How to post-process an HDF5 file][post-process-an-hdf5-file]
for more information on how to generate and load such a file.

!!! note
    The `execute_post` function also handles previously executed instances of `MDOScenario`
    or `OptimzationProblem`.



### 2. Plot the objective history




```python
execute_post(
    "sobieski_mdf_scenario.h5",
    settings_model=BasicHistory_Settings(
        variable_names=["y_4"],
        save=False,
        show=True,
    ),
)
```

### 3. Plot the constraint history




```python
execute_post(
    "sobieski_mdf_scenario.h5",
    settings_model=BasicHistory_Settings(
        variable_names=["g_1", "g_2", "g_3"],
        save=False,
        show=True,
    ),
)
```

## Summary

[BasicHistory][gemseo.post.basic_history.BasicHistory] plots the iteration-by-iteration
history of any problem function.
Use `variable_names` to select the functions to display.
Pass any HDF5 file from a previously executed scenario to
[execute_post()][gemseo.execute_post].
